```python
import wikipedia
from transformers import pipeline

# Step 1: Load models
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def wiki_qa_detailed(query, topic):
    """
    query: the question you want to ask
    topic: Wikipedia article to fetch (e.g., 'French Revolution', 'Theory of Relativity')
    """
    try:
        # Fetch Wikipedia page
        page = wikipedia.page(topic)
        context = page.content[:4000]   # limit length for performance

        # Step 2: Extract a short answer
        result = qa_pipeline(question=query, context=context)
        short_answer = result['answer']

        # Step 3: Summarize relevant context into a detailed answer
        summary = summarizer(context[:1500], max_length=120, min_length=60, do_sample=False)
        detailed_answer = summary[0]['summary_text']

        return short_answer, detailed_answer

    except Exception as e:
        return None, f"Error: {str(e)}"


# -------------------------
# 🔎 Example Usage
# -------------------------

# Historical query
q1 = "When did the French Revolution begin?"
short, detailed = wiki_qa_detailed(q1, "French Revolution")
print(f"Q: {q1}\nShort Answer: {short}\nDetailed Answer: {detailed}\n")

# Scientific query
q2 = "Who developed the theory of relativity?"
short, detailed = wiki_qa_detailed(q2, "Theory of Relativity")
print(f"Q: {q2}\nShort Answer: {short}\nDetailed Answer: {detailed}\n")

# Gandhi query
q3 = "What method did Gandhi use in his campaigns?"
short, detailed = wiki_qa_detailed(q3, "Mahatma Gandhi")
print(f"Q: {q3}\nShort Answer: {short}\nDetailed Answer: {detailed}\n")
```
